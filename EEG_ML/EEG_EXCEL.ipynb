# Excel 파일을  CSV 파일로 만드는 함수 만들기

```python
data_dir = r"C:\Users\BraveHong\Documents\GitHub\EEG_data"
```

```python
def EEG_data_csv(data_dir,normalization = 'Abs'):
    file_list = os.listdir(data_dir)
    file_list.sort()
    print("normalization 종류" ,file_list)
    print( "선택한 normalization : ", normalization)
    EEG_dir= data_dir+"\\"+normalization
    EEG_list= os.listdir( EEG_dir)
    EEG_list.sort()
    
    
    
    #각 EEG 엑셀 파일들을 각 변수로 지정 
    alpha_file = EEG_dir+"\\"+EEG_list[0]
    beta_file = EEG_dir+"\\"+EEG_list[1]
    delta_file = EEG_dir+"\\"+EEG_list[2]
    gamma_file =  EEG_dir+"\\"+EEG_list[3]
    theta_file = EEG_dir+"\\"+EEG_list[4]
    
    
    
    #sheet 이름 변수
    alpha_sheet_name = []
    wb = load_workbook(alpha_file)
    for ws in wb.worksheets :
        alpha_sheet_name.append(ws.title)

    
    
    beta_sheet_name = []
    wb = load_workbook(beta_file)
    for ws in wb.worksheets :
        beta_sheet_name.append(ws.title)

        

    delta_sheet_name = []
    wb = load_workbook(delta_file)
    for ws in wb.worksheets :
        delta_sheet_name.append(ws.title)

    

    gamma_sheet_name = []
    wb = load_workbook(gamma_file)
    for ws in wb.worksheets :
        gamma_sheet_name.append(ws.title)

    
    theta_sheet_name = []
    wb = load_workbook(theta_file)
    for ws in wb.worksheets :
        theta_sheet_name.append(ws.title)

    
    #엑셀 시트별로 csv 파일 생성하기 위해 디렉토리 생성 및 csv 파일 생성
    
    csv_dir = "./EEG_csv_data"
    if not os.path.isdir(csv_dir):
        os.mkdir(csv_dir)
    
    Abs_dir ="./EEG_csv_data/Abs"
    Relative_dir = "./EEG_csv_data/Relative"
    
    if not os.path.isdir(Abs_dir):
        os.mkdir(Abs_dir)
    if not os.path.isdir(Relative_dir):
        os.mkdir(Relative_dir)
    
  
    
    for i in range(len(alpha_sheet_name)):
        alpha_dir = csv_dir+"\\"+normalization+"\\"+"alpha"+"\\"+alpha_sheet_name[i]+".csv"
        alpha_dir1 = csv_dir+"\\"+normalization+"\\"+"alpha"
        if not os.path.isdir(alpha_dir1):
            os.mkdir(alpha_dir1)
        alpha_sheet_csv = pd.read_excel(alpha_file,sheet_name=alpha_sheet_name[i])
        alpha = alpha_sheet_csv.to_csv(alpha_dir,mode='w')
    
    
    
    for i in range(len(beta_sheet_name)):
        beta_dir  = csv_dir+"\\"+normalization+"\\"+"beta"+"\\"+beta_sheet_name[i]+".csv"
        beta_dir1  = csv_dir+"\\"+normalization+"\\"+"beta"
        if not os.path.isdir(beta_dir1):
            os.mkdir(beta_dir1)
        beta_sheet_csv =pd.read_excel(beta_file,sheet_name=beta_sheet_name[i])
        beta = beta_sheet_csv.to_csv(beta_dir,mode='w')
   


    for i in range(len(delta_sheet_name)):
        delta_dir = csv_dir+"\\"+normalization+"\\"+"delta"+"\\"+delta_sheet_name[i]+".csv"
        delta_dir1 = csv_dir+"\\"+normalization+"\\"+"delta"
        if not os.path.isdir(delta_dir1):
            os.mkdir(delta_dir1)
        delta_sheet_csv =pd.read_excel(delta_file,sheet_name=delta_sheet_name[i])
        delta = delta_sheet_csv.to_csv(delta_dir,mode='w')

        
    for i in range(len(gamma_sheet_name)):
        gamma_dir = csv_dir+"\\"+normalization+"\\"+"gamma"+"\\"+gamma_sheet_name[i]+".csv"
        gamma_dir1 = csv_dir+"\\"+normalization+"\\"+"gamma"
        if not os.path.isdir(gamma_dir1):
            os.mkdir(gamma_dir1)
        gamma_sheet_csv =pd.read_excel(gamma_file,sheet_name=gamma_sheet_name[i])
        gamma = gamma_sheet_csv.to_csv(gamma_dir,mode='w')
        
        
        
    for i in range(len(theta_sheet_name)):
        theta_dir = csv_dir+"\\"+normalization+"\\"+"theta"+"\\"+theta_sheet_name[i]+".csv"
        theta_dir1 = csv_dir+"\\"+normalization+"\\"+"theta"
        if not os.path.isdir(theta_dir1):
            os.mkdir(theta_dir1)
        theta_sheet_csv =pd.read_excel(theta_file,sheet_name=theta_sheet_name[i])
        theta = theta_sheet_csv.to_csv(theta_dir,mode='w')
        
        
    
    
```
